## 1. Make classification data and get it ready


```python
import sklearn # machine learning library

from sklearn.datasets import make_circles

import pandas as pd

import matplotlib.pyplot as plt

```

```python
# make 1000 samples

n_samples = 1000

# create_circles
# noise = randomness
X, y = make_circles(n_samples, 
                    noise =0.03, 
                    random_state=42)
# make circles gibt uns ein sample dataset.
# X = matrix aus je 2 daten werten
# y = Klassifizieren 1 oder 0
len(X), len(y)
print(f"first 5 Samples of X{ X[:5]}")
print(f"first 5 Samples of y{ y[:5]}")
```

```python
circles = pd.DataFrame({"X1": X[:,0],   #everythin in 0
                        "X2": X[:,1],
                        "label": y
                        })

circles.head(10) # wir könne damit wenig anfangen

```

```python
# können wir das Muster graphish erkennen?
plt.scatter(x=X[:,0],
            y=X[:,1],
            c = y,
            
            cmap =plt.cm.RdYlBu)

plt.legend()
plt.plot()

# jeder punkt wir entwer y = 1 oder y= 0 zugeordnet. 
# jede dieser Kategorien form einen Ringe, einen äußeren ode inneren Ring
```
